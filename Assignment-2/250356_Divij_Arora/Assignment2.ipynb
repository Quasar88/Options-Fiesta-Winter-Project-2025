```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def buy_call_payoff(S, K, P):
    return np.maximum(0, S - K) - P

def sell_call_payoff(S, K, P):
    return P - np.maximum(0, S - K)

def buy_put_payoff(S, K, P):
    return np.maximum(0, K - S) - P

def sell_put_payoff(S, K, P):
    return P - np.maximum(0, K - S)
```

```python
# K = int(input("Enter the strike price: "))
# P = int(input("Enter the premium: "))
# A = int(input("Enter the minimum spot price: "))
# B = int(input("Enter the maximum spot price:"))
K = 100       # Strike Price [cite: 8]
P = 5         # Premium [cite: 9]
# We generate the whole range at once using NumPy
# Range from K-50 to K+30 as per instructions [cite: 10]
spot_range = np.linspace(K - 50, K + 30, 100)
# spot_range = np.linspace(A,B, 100)

# --- Generate Payoffs ---
bc_payoff = buy_call_payoff(spot_range, K, P)
sc_payoff = sell_call_payoff(spot_range, K, P)
bp_payoff = buy_put_payoff(spot_range, K, P)
sp_payoff = sell_put_payoff(spot_range, K, P)
```

```python
plt.plot(spot_range, bc_payoff, color='green')
plt.axhline(0, color='black')
plt.title('Buy Call')
plt.xlabel('Spot Price')
plt.ylabel('Profit / Loss')
```

```python
plt.plot(spot_range, sc_payoff, color='red')
plt.axhline(0, color='black', lw=1)
plt.title('Sell Call Payoff')
plt.xlabel('Spot Price')
plt.ylabel('Profit / Loss')
```

```python
plt.plot(spot_range, bp_payoff, color='blue')
plt.axhline(0, color='black', lw=1)
plt.title('Buy Put Payoff')
plt.xlabel('Spot Price')
plt.ylabel('Profit / Loss')
```

```python
plt.plot(spot_range, sp_payoff, color='orange')
plt.axhline(0, color='black', lw=1)
plt.title('Sell Put Payoff')
plt.xlabel('Spot Price')
plt.ylabel('Profit / Loss')
```
