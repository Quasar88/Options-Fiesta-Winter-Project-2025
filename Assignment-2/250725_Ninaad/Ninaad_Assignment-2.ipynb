```python
import matplotlib.pyplot as plt
import numpy as np 
import pandas as pd
```

```python
# Inputs :
 
K = 100
P = 10
"""
K = float(input("Enter strike price K: "))
P = float(input("Enter premium P: "))
"""
```

```python
# Range of spot prices
S_min = K - 50
S_max = K + 30
S = np.linspace(S_min, S_max, 200)
```

```python
# Payoffs : 
buy_call = np.maximum(0, S - K) - P 
 
sell_call = P - np.maximum(0, S - K)
 
buy_put = np.maximum(0, K - S) - P

sell_put = P - np.maximum(0, K - S)
```

```python
# Plotting : 
fig, axs = plt.subplots(2, 2, figsize=(10, 8), sharex=True, sharey=True)

# Buy Call
axs[0, 0].plot(S, buy_call, color = 'green')
axs[0, 0].axhline(0, color = 'black', linewidth = 0.8)
axs[0, 0].axvline(K, color = 'gray', linestyle = '--', linewidth=0.8) #indicating strike price with "--"
axs[0, 0].set_title("Buy Call")
axs[0, 0].set_xlabel("Spot price")
axs[0, 0].set_label("Profit/Loss")


# Sell Call
axs[0, 1].plot(S, sell_call, color = 'red')
axs[0, 1].axhline(0, color = 'black', linewidth = 0.8)
axs[0, 1].axvline(K, color = 'gray', linestyle = '--', linewidth = 0.8)
axs[0, 1].set_title("Sell Call")
axs[0, 1].set_xlabel("Spot price")
axs[0, 1].set_label("Profit/Loss")


# Buy Put
axs[1, 0].plot(S, buy_put, color = 'red')
axs[1, 0].axhline(0, color = 'black', linewidth = 0.8)
axs[1, 0].axvline(K, color = 'gray', linestyle = '--', linewidth = 0.8)
axs[1, 0].set_title("Buy Put")
axs[1, 0].set_xlabel("Spot price")
axs[1, 0].set_label("Profit/Loss")


# Sell Put
axs[1, 1].plot(S, sell_put, color = 'green')
axs[1, 1].axhline(0, color = 'black', linewidth=0.8)
axs[1, 1].axvline(K, color = 'gray', linestyle = '--', linewidth = 0.8) 
axs[1, 1].set_title("Sell Put")
axs[1, 1].set_xlabel("Spot price")
axs[1, 1].set_label("Profit/Loss")


for ax in axs.flat:
    ax.grid(True)

plt.show()
```
