```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
K= float(input("Enter the Strike Price: ")) #Input Strike Price
P= float(input("Enter the Premium: "))      #Input Premium

start = float(input("Enter the lower limit of Spot Price: ")) #Lowest value of spot price range
stop =  float(input("Enter the upper limit of Spot Price: ")) #Highest value of spot price range

option_type = input("Enter the type of option(call/put): ").strip().lower()

S = np.arange(start, stop, 1)

if option_type=="call":
    payoff = np.maximum(S-K,0)    #calculating the payoff
    profit_call_buy = payoff - P  #calculating the profit of the buyer 
    profit_call_sell = P-payoff   #calculating the profit of the seller

    fig, ax = plt.subplots(1,2)
    ax[0].plot(S, profit_call_buy, c='r')
    ax[0].plot([start, stop], [0,0], ls = '--')    #plotting down the x axis
    ax[0].set_xlabel("Spot Price")
    ax[0].set_ylabel("Profit")
    ax[0].set_title("Buying a call option")

    ax[1].plot(S, profit_call_sell, c='r')
    ax[1].plot([start, stop], [0,0], ls = '--')    #plotting down the x axis
    ax[1].set_xlabel("Spot Price")
    ax[1].set_ylabel("Profit")
    ax[1].set_title("Selling a call option")

    breakeven_x = K + P   #plotting the breakeven point
    breakeven_y = 0
    ax[0].plot(breakeven_x, breakeven_y, marker='o', color='black')  
    ax[1].plot(breakeven_x, breakeven_y, marker='o', color='black')


if option_type=="put":
    payoff = np.maximum(K-S,0)    #calculating the payoff
    profit_put_buy = payoff - P   #calculating the profit of the buyer  
    profit_put_sell = P-payoff    #calculating the profit of the seller

    fig, ax = plt.subplots(1,2)
    ax[0].plot(S, profit_put_buy, c='r')
    ax[0].plot([start, stop], [0,0], ls = '--')    #plotting down the x axis
    ax[0].set_xlabel("Spot Price")
    ax[0].set_ylabel("Profit")
    ax[0].set_title("Buying a put option")

    ax[1].plot(S, profit_put_sell, c='r')
    ax[1].plot([start, stop], [0,0], ls = '--')    #plotting down the x axis
    ax[1].set_xlabel("Spot Price")
    ax[1].set_ylabel("Profit")
    ax[1].set_title("Selling a put option")

    breakeven_x = K - P   #plotting the breakeven point
    breakeven_y = 0
    ax[0].plot(breakeven_x, breakeven_y, marker='o', color='black')
    ax[1].plot(breakeven_x, breakeven_y, marker='o', color='black')

    plt.show()
 
```

```python
# OBSERVATIONS:

# 1. Buying a Call
# Loss is limited to the premium when spot price < strike price.
# Profit increases when spot price > strike price.
# Breakeven: K+P
# Unlimited profit, limited loss.

# 2. Selling a Call
# Profit is limited to the premium received.
# Loss increases when spot price rises.
# Breakeven: K+P
# Limited profit, unlimited loss.

    
# 3. Buying a Put
# Profit increases when spot price falls below strike price.
# Loss is limited to the premium paid.
# Breakeven: K−P

# 4. Selling a Put
# Profit is limited to the premium received.
# Loss increases as spot price falls.
# Breakeven: K−P

# In general the profit curves of for buyer and seller are mirror images of each other.
```
