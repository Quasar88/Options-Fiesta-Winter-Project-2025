```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
k=int(input("Enter strike price: "))
p=int(input("Enter premium: "))
low, high = map(int, input("Enter range of spot prices separated by space: ").split())
otype = input("Enter type of option: ").lower()
```

```python
plt.subplot(2,2,1)
s=np.arange(low,high+1)
profit=np.maximum(0,s-k)-p
plt.plot(s,profit)
plt.axhline(y=0, color='black', linewidth=1, linestyle='--')
plt.xlabel("Spot Price")
plt.ylabel("Profit/Loss")
plt.title("Buying a Call Option")

plt.subplot(2,2,2)
s=np.arange(low,high+1)
profit = p - np.maximum(0,s-k)
plt.plot(s,profit)
plt.axhline(y=0, color='black', linewidth=1, linestyle='--')
plt.xlabel("Spot Price")
plt.ylabel("Profit/Loss")
plt.title("Selling a Call Option")

plt.subplot(2,2,3)
s=np.arange(low,high+1)
profit=np.maximum(0,k-s)-p
plt.plot(s,profit)
plt.axhline(y=0, color='black', linewidth=1, linestyle='--')
plt.xlabel("Spot Price")
plt.ylabel("Profit/Loss")
plt.title("Buying a Put Option")

plt.subplot(2,2,4)
s=np.arange(low,high+1)
profit = p - np.maximum(0,k-s)
plt.plot(s,profit)
plt.axhline(y=0, color='black', linewidth=1, linestyle='--')
plt.xlabel('Spot Price')
plt.ylabel("Profit/Loss")
plt.title("Selling a Put Option")

plt.tight_layout()
plt.show()
```

# Option Greeks

1) **Delta**


*   Tells us how much the option price changes when the original stock price changes by 1 dollar
*   If delta is 0.5 then a 1 dollar increase in price of the original stock will increase the option value by 0.5

*   Range: Calls:[0,1] Puts:[-1,0]







2) **Gamma**

*   Tells us how much delta changes if stock price changes by 1 dollar
*   Tells us how fast delta is changing. High gamma means high risk and delta is changing rapidly

*   Range: always positive






3) **Theta**

*   Tells us how much option value is lost each day as time passes
*   It is non linear and accelerates rapidly in the last days


*   Range: Negative for buyers(Long), Positive for buyers(Short)





4. **Vega**


*   Tells us the change in option price when implied volatality changes by 1%
*   Higher IV means higher price since vega increases the premium

*   Range: Always Positive





5. **Rho**


*   Tells us the change in price of option when interest rate changes by 1%
*   More relevent for long dated options



*   Range: Calls: positive(more rate means increase in value) Puts:negative(more rate means decrease in value)
