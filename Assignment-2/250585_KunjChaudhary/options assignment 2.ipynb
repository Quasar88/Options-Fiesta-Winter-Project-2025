```python
## strike_price = k,premium = p and spot price = s
## possibly the graph you're seeing is of call, that is because i've given user input for call and put and the test case used is call.
import numpy as np
import matplotlib.pyplot as plt

k = float(input("Enter Strike Price:"))
p = float(input("Enter Premium:"))
option_type = input("Enter call or put:").lower()

s = np.linspace(max(0, k - 50), k + 50, 1000)

if option_type == "call": #case 1 where user inputs call
    # Buy and sell
    buy_payoff = np.maximum(s - k, 0) - p #payoff for buy in call option
    sell_payoff = p - np.maximum(s - k, 0) #payoff for sell in call option
    
    plt.plot(s, buy_payoff, label="Buy Call")
    plt.plot(s, sell_payoff, label="Sell Call")
    plt.ylabel("Call Option Payoff")
    plt.title("Call Option Payoff vs Spot Price")

elif option_type == "put": #case 2 where user inputs call
    # Buy and sell
    buy_payoff = np.maximum(k - s, 0) - p #payoff for buy in put option
    sell_payoff = p - np.maximum(k - s, 0) #payoff for sell in put option
    
    plt.plot(s, buy_payoff, label="Buy Put")
    plt.plot(s, sell_payoff, label="Sell Put")
    plt.ylabel("Put Option Payoff")
    plt.title("Put Option Payoff vs Spot Price")


plt.xlabel("Spot Price")
plt.legend()
plt.show()
```


## what i observe in the plot is that the graphs for a particuar option(call or put) are just flipped by 180 degress and an overall symmetricity is observed, after a certain point the buyer will always make a profit and the seller will continue to make loss while the buyer's loss is limited to the premium paid.

## now about the greeks
## DELTA - this is basically the rate of change of premium wrt to the change in strike price, it is btwn 0 to 1 for call and 0 to -1 for put
## GAMMA - it is the rate of change in delta, if suppose premium increases delta increases and similarly gamma increases,it approaches 0 for ITM and OTM while it is maximum at ATM
## THETA - this is like the daily value which depreciates of premium, it increases when the date of expiry approaches. if you're buying then the longer you wait the premium decreases so theta is negative when buying and vice versa
## VEGA - it implies volatility, if the volatility changes vega multiplies by the change and is added to the premium.
## RHO - it is the amount of change in premium when interest changes by 1%, +ve for calls and -ve for puts.
