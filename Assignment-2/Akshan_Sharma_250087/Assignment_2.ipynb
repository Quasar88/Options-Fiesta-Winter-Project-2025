```python
from matplotlib import pyplot as plt
import numpy as np
```

```python
K = float(input("Enter Strike Price: "))
P = float(input("Enter the Premium: "))
mi, ma = map(float, input("Enter the space separated minimum and maximum spot price: ").split())
option = {"call":1, "put":-1}[input("Enter your option (call / put): ").lower()]

x = np.arange(mi, ma, 0.1)
y = np.array([max(0, (s-K)*option) - P for s in x])

#Buying
plt.plot(x, y)
plt.show()

#Selling
plt.plot(x, -y)
plt.show()
```
