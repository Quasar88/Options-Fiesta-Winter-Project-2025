```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

## Basic Variables Part

```python
K = float(input("Input the Strike Price: "))
P = float(input("Input the Premium: "))

SL = float(input("Input the lower limit of Spot Price: "))
while (SL >= K):
    SL = float(input("Input the lower limit of Spot Price: "))
    
SU = float(input("Input the upper limit of Spot Price: "))
while  (SU <= K):
    SU = float(input("Input the upper limit of Spot Price: "))
    
T = input("'Call' or 'Put': ")
while (T != 'Call' and T != 'Put'):
    T = input("Invalid input. Enter 'Call' or 'Put': ")
```

## Payoff Part

```python
X = np.linspace(SL, SU, 100)

Call_Buy = np.maximum(0, X - K) - P
Call_Sell = P - np.maximum(0, X - K)
Put_Buy = np.maximum(0, K - X) - P
Put_Sell = P - np.maximum(0, K - X)
```

## Graph Part

```python
plt.plot(X, Call_Buy, label="Buy Call", color = "green")
plt.plot(X, Call_Sell, label="Sell Call", color = "red")

plt.axvline(K + P, linestyle="--", label="Breakeven")
plt.axhline(0, color = "black")

plt.title("Payoffs for Call Option")
plt.xlabel("Spot Price")
plt.ylabel("Profit / Loss")

plt.legend()
plt.grid()
plt.show()
```

```python
plt.plot(X, Put_Buy, label="Buy Put", color = "green")
plt.plot(X, Put_Sell, label="Sell Put", color = "red")

plt.axvline(K - P, linestyle="--", label="Breakeven")
plt.axhline(0, color = "black")

plt.title("Payoffs for Put Option")
plt.xlabel("Spot Price")
plt.ylabel("Profit / Loss")

plt.legend()
plt.grid()
plt.show()
```

## Greeks
#### Delta is the change in premium per unit change in the stock.
#### Gamma is the change in delta per unit change in the stock.
#### Theta represents time decay; change in the premium per day.
#### Vega is the change in premium per unit change in implied volatility.
#### Rho is the change in premium per cent change in interest rate.
